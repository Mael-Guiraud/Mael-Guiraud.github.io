```python
import graphviz
class Node:
   
    def __init__(self, data,left=None,right=None):
        self.data = data
        self.left = left
        self.right = right
        self.d= graphviz.Digraph()
    
    def __str__(self):
        return str(self.data)
    
    def print(self):
        if(self.right == None and self.left==None):
            self.d.node(str(self.data),str(self.data))
            print('pas de fils')
        else:
            self.print_node(self.d)
            print('fils')
    def print_node(self,d):
        if(self.left):
            d.edge(str(self.data),str(self.left.data))
            self.left.print_node(d)
        if(self.right):
            d.edge(str(self.data),str(self.right.data))
            self.right.print_node(d)
        
    def insert(self,data):
        if self.data == data:
            return
        if self.data > data:
            if self.left == None:
                self.left = Node(data)
            else:
                self.left.insert(data)
        if self.data < data:
            if self.right == None:
                self.right = Node(data)
            else:
                self.right.insert(data)
    
n=Node(l[0])
for i in range(1,len(l)):
    n.insert(l[i])
n.print()
n.d
```
