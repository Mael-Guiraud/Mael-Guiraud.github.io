```python
import matplotlib.pyplot as plt
import matplotlib.axes
import numpy as np



labs = ['SDA','HP','SW','RVLF','PDC','IJ']
men_means = [617,284,213,49,22,14]

x = np.arange(len(labs))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=[10,6])
rects1 = ax.bar(x , men_means, width)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Nombre de votes')

ax.set_title('Résultats le 26/11/2021 à 13h17')
ax.set_xticks(x)
ax.set_xticklabels(labs)

ax.legend()

ax.bar_label(rects1, padding=3)

fig.tight_layout()

plt.show()



# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'SDA','HP','SW','RVLF','PDC','IJ'
sizes = [617,284,213,49,22,14]
explode = (0.1, 0,0,0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
